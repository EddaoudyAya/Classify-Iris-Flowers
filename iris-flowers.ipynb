# Classify Iris Flowers

Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

loading the dataset

```python
columns=['sepal length','sepal width','petal length','petal width','class']
df=pd.read_csv('iris.csv.csv',names=columns)
df = df.iloc[1:]
df.head()
```

# visualization

```python
df.describe()
```

```python
numeric_cols = ['sepal length', 'sepal width', 'petal length', 'petal width']
df[numeric_cols] = df[numeric_cols].astype(float)

# Create pair plot
sns.pairplot(df, hue='class')

# Show the plot
plt.show()
```

seperating input and output columns

```python
data=df.values
x=data[:,0:4]
y=data[:,4]
```

```python
x
```

```python
y
```

splitting the data into training and testing

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# MODELS

Model 1:support vector machine algorithm

```python
model_svc=SVC()
model_svc.fit(x_train,y_train)
```

```python
prediction1=model_svc.predict(x_test)
print(accuracy_score(y_test,prediction1))
```

Model 2: logistic regression

```python
model_lr=LogisticRegression()
model_lr.fit(x_train,y_train)
```

```python
prediction2=model_lr.predict(x_test)
print(accuracy_score(y_test,prediction2))
```

Model 3:decision tree classifier

```python
model_dtc=DecisionTreeClassifier()
model_dtc.fit(x_train,y_train)
```

```python
prediction3=model_dtc.predict(x_test)
print(accuracy_score(y_test,prediction3))
```

```python
print(classification_report(y_test,prediction2))
```

# Testing

```python
x_new=np.array([[3,2,1,0.2],[4.9,2.2,3.8,1.1]])
prediction=model_lr.predict(x_new)
print('prediction of species{}'.format(prediction))
```
